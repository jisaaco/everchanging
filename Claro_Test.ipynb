### 2 SEGMENTACIÓN DE CLIENTES RFM

### Calculando Recency, Frequency, Monetary Value

```python
ClaroJoin['ORDERDATE'] = pd.to_datetime(ClaroJoin['ORDERDATE'])
ClaroJoinD = ClaroJoin[(ClaroJoin['ORDERDATE'] >= '2003-1-1') & (ClaroJoin['ORDERDATE'] <= '2005-5-31')] #Para el rango de fechas sugerido
ClaroRFM_FR = ClaroJoinD.groupby('ID_Cliente').size().to_frame(name = 'Frequency')
ClaroRFM_MV = ClaroJoinD.groupby('ID_Cliente')['SALES'].sum().to_frame(name = 'Monetary Value')
ClaroRFM_LP = ClaroJoinD.sort_values('ORDERDATE').groupby('ID_Cliente').tail(1).set_index('ID_Cliente')['ORDERDATE'].to_frame(name = 'Last Purchase').sort_index()
ClaroRFM = ClaroRFM_LP.join(ClaroRFM_FR).join(ClaroRFM_MV)
ClaroRFM.reset_index(level=0, inplace=True)
Today = pd.to_datetime('2021-11-10')
ClaroRFM['Recency'] = Today - ClaroRFM['Last Purchase']
ClaroRFM['Recency'] = ClaroRFM['Recency'].dt.days
ClaroRFM.drop('Last Purchase', axis = 'columns', inplace = True)
ClaroRFM
```

### Calculando Scores

```python
ClaroRFM['Frequency Score'] = pd.cut(ClaroRFM['Frequency'],
                                     bins = [-1,
                                           np.percentile(ClaroRFM['Frequency'], 20),
                                           np.percentile(ClaroRFM['Frequency'], 40),
                                           np.percentile(ClaroRFM['Frequency'], 60),
                                           np.percentile(ClaroRFM['Frequency'], 80),
                                           ClaroRFM['Frequency'].max()],
                                     labels = [1, 2, 3, 4, 5]).astype("int")

ClaroRFM['Monetary Value Score'] = pd.cut(ClaroRFM['Monetary Value'],
                                     bins = [-1,
                                           np.percentile(ClaroRFM['Monetary Value'], 20),
                                           np.percentile(ClaroRFM['Monetary Value'], 40),
                                           np.percentile(ClaroRFM['Monetary Value'], 60),
                                           np.percentile(ClaroRFM['Monetary Value'], 80),
                                           ClaroRFM['Monetary Value'].max()],
                                     labels = [1, 2, 3, 4, 5]).astype("int")

ClaroRFM['Recency Score'] = pd.cut(ClaroRFM['Recency'],
                                     bins = [-1,
                                           np.percentile(ClaroRFM['Recency'], 20),
                                           np.percentile(ClaroRFM['Recency'], 40),
                                           np.percentile(ClaroRFM['Recency'], 60),
                                           np.percentile(ClaroRFM['Recency'], 80),
                                           ClaroRFM['Recency'].max()],
                                     labels = [5, 4, 3, 2, 1]).astype("int")
```

```python
ClaroRFM
```

### 4 MODELO DE RECOMENDACIÓN

**Se toma el número de orden, código de producto y la cantidad ordenada. Y se pivotea convirtiendo a los productos en columnas con valor el número de piezas compradas por orden (asignando ceros a los productos no pedidos).**

```python
ClaroRS = ClaroJoinD[['ORDERNUMBER', 'QUANTITYORDERED', 'PRODUCTCODE']].sort_values('ORDERNUMBER')
ClaroRS
```

```python
ClaroMatrix = ClaroRS.pivot(index = 'ORDERNUMBER', columns = 'PRODUCTCODE', values = 'QUANTITYORDERED')
ClaroMatrix.fillna(0, inplace = True)
ClaroMatrix
```

Se calcula una matriz de distancias entre productos

```python
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform

matrix = pd.DataFrame(
    squareform(pdist(ClaroMatrix)),
    columns = ClaroMatrix.index,
    index = ClaroMatrix.index
)

matrix
```

**Así, es posible elegir los tres productos más "cercanos" al producto del cual se solicitaron más piezas como sugerencias de compra, y hacer lo mismo para el cual se solicitaron menos piezas. Por ejemplo al comprar el producto 10100:**

```python
RecCode = matrix.sort_values(by = 10100, ascending = True)[10100].head(4).to_frame('Distancia').drop(10100)
RecCode
```

**Para la recomendación se puede cargar una función como la siguiente para que solo arroje los productos recomendados. Con el dato del nombre del producto se puede reemplazar su código por dicho nombre.**

```python
def Recomienda(ordnum):
    RecCode = matrix.sort_values(by = ordnum, ascending = True)[ordnum].head(4).to_frame(name = 'Drop').drop(ordnum)
    RecCode.reset_index(level=0, inplace=True)
    RecCode.drop('Drop', axis='columns', inplace=True)
    RecCode.rename(columns = {'ORDERNUMBER':'Recomendación'}, inplace = True)
    return RecCode
```
